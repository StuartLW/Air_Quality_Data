The core function is importUKAQ which will produce a single merged dataframe with data downloaded from the air quality networks.  It uses translation fo the R data directly into python with code ported from the R functions developed by the [open air project](https://github.com/openair-project/openair]) who have done all the heavy lifting here.  <br>
The core function requires site ids for the target sites, years or ranges of years;  the source network; the desired data frequeny and optionally target pollutants. There are also a range of formatting and output options.  The downloaded datasets can also be stored in hdf5 file format for future retrieval.  Note that selection of datafrequancy file and pollutants happens after download and so makes no difference to download speed.  In addition all the downloaded data are saved in the hdf file if specified.

```python
# import all polutant data from site LED6 from the aurn network for 2024 and keep the hourly data
df = importUKAQ('LED6',2024, 'aurn', data_type='hourly', pollutant='any' )
df.head()
```

```python
# import NO and NO2 data from sites LED6 and LEED for 2020 and 2024 from the 'aurn' network
df = importUKAQ(['LED6', 'LEED'],[2020,2024], 'aurn', data_type='hourly', pollutant=['NO', 'NO2'] )
df.head()
```

```python
# import daily NO and NO2 data from sites LED6 2020:2024 inclusive from the 'aurn' network
df = importUKAQ('LED6','2020:2024', 'aurn', data_type='daily', pollutant=['NO', 'NO2'] )
df.head()
```

```python
# get 2020 data from all sites that record all of O3, NO and ETHANE
df = importUKAQ('all', 2020, 'aurn', data_type='hourly', pollutant=['O3','NO','ETHANE'], write_raw_toHDF='o3_no_ethane_2020.h5' )
df.head()
```

```python
# reread the previously written hdf5 file and take the dialy data for any pollutatnts.  Note as the save is done on all the downloaded
# data then we do not need to redownload it
df = importUKAQ('all', 2020, 'o3_no_ethane_2020.h5', data_type='daily', pollutant='any' )
df.head()
```

```python
# more complex multisource queries are also possible. here import the leed and led6 sites from aurn and combine with all the sites in the hdf5 file
# report daily values and convert to narrow/ long format
df = importUKAQ([['leed', 'led6'],'all'], 2020, ['aurn', 'o3_no_ethane_2020.h5'], data_type='daily', pollutant='any', to_narrow=True)
df.loc[1825:1835]
```
